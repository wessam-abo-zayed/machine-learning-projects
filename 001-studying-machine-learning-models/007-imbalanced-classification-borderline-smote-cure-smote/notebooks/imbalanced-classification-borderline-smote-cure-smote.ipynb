<center><b><h2>STOME</h2></b></center>

- Synthetic Minority Oversampling Technique (SMOTE) is a well-known oversampling technique for imbalanced data classification problems at the data preprocessing step.
- This method is motivated by a successful technique in handwritten character recognition that creates the extra training data by performing specific operations, such as rotating handwritten characters.
- Frstly selects a minority instance randomly and finds the k-nearest neighbors of that instance. A synthetic instance is then created at a randomly selected point on the line connecting the instance and one random instance among the k neighbors in the feature space.

- ### How does SMOTE work ?
<br>
 - The minority class is over-sampled by taking each minority class sample and introducing synthetic examples along the line segments joining any/all of the k minority class nearest neighbors.
 <br><br>
 - Synthetic samples are generated in the following way: Take the diﬀerence between the feature vector (sample) under consideration and its nearest neighbor. Multiply this diﬀerence by a random number between 0 and 1, and add it to the feature vector under consideration. This causes the selection of a random point along the line segment between two speciﬁc features. This approach eﬀectively forces the decision region of the minority class to become more general.
 <br><br>
 - For a minority instance, the number of randomly chosen neighbors among its k-nearest neighbors depends on the amount of oversampling instances required. For example, with k = 5, if the needed amount is three times the size of the original data, three neighbors out of five nearest neighbors will be chosen, and one synthetic instance is generated on each line connecting the minority instance and the selected neighbors. The following steps can be used to create a synthetic instance:<br><br>
     - Pick an instance from the minority class randomly.<br>
     - Find the k-nearest neighbors of the current instance.<br>
     - Take the difierence between the current instance and one of the neighbors, then multiply the difierence by a random number gap between 0 and 1.
     - Add the result from the previous step to the current instance and create a new synthetic instance.

After a neighbor xneighbor is randomly selected, the synthetic instance xsynth is generated along the line joining xi and xneighbor.

<center><b><h2>Borderline-SMOTE</h2></b></center>

- In the training process, Borderline-SMOTE algorithms attempt to learn the borderline of each class, where these borderline instances and the ones nearby are more likely to be misclassified than the ones far from the borderline. In Borderline-SMOTE, all instances in the minority class are divided into three groups: NOISE, DANGER, and SAFE:
<ul>
    <li><b>NOISE:</b> instances are rare, probably incorrect, located in areas occupied by the majority
class instances.</li>
    <li><b>DANGER:</b> instances are placed in the area near class boundaries and typically overlap
with majority instances.</li>
    <li><b>SAFE:</b> instances are relatively easier to recognize, and they are the main representatives
of the minority class.</li>
    
<ul>

 <img src="images/p14.png" width="400">

<br>
- After all instances of the minority class are categorized, synthetic instances are then created along the line between DANGER instances and their nearest neighbors. <center><b><h2>CURE-SMOTE</h2></b></center>

- the problem of classifying imbalanced data has attracted increasing attention. Imbalanced data sets generally refer to data that is distributed unevenly among different categories where the data in the smaller category is far less prevalent than data in the larger category. The Imbalance Ratio (IR) is defined as the ratio of the number of minor class samples to the number of major class samples. Therefore, imbalanced data causes the training set for each decision tree to be imbalanced during the first “random” procedure. The classification performance of traditional RF on imbalanced data sets is even worse than that of SVMs

- The general idea of the CURE-SMOTE algorithm is as follows: cluster the samples of the minor class using CURE, remove the noise and outliers from the original samples, and, then, generate artificial samples randomly between representative points and the centre point. The implementation steps of the CURE-SMOTE algorithm are as follows:

    <li>1- Normalize the dataset, extract the minor class
samples, X, and calculate the distance dist among them.

- During the clustering process of the CURE-SMOTE algorithm, noisy points must be removed because they are far away from the   normal points, and they hinder the merge speed in the corresponding class. 
<br><br>
- When clustering is complete, the clusters containing only a few samples are also deemed to be noisy points. For the sample points after clustering, the interpolation can effectively prevent generalization and preserve the original distribution attributes of the data set. 
<br><br>
- In the interpolation formula, Xi is replaced by the representative points; consequently, the samples are generated only between the representative samples and the samples in the original minor class, which effectively avoids the influence of boundary points. 
<br><br>
- The combination of the clustering and merge operations serves to eliminate the noise points at the end of the process and reduce the complexity because there is no need to eliminate the farthest generated artificial samples after the SMOTE algorithm runs. Moreover, all the termination criteria such as reaching the pre-set number of clusters, the number of representative samples, or the distance threshold, avoid setting the k value of the original SMOTE algorithm and, thus, reduce the instability of the proposed algorithm.

<h3>Python Implimintation:</h3> (Credit Card Fraud Detection)

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from time import time
from collections import Counter
from imblearn.over_sampling import SMOTE
from sklearn import metrics
```

```python
df = pd.read_csv('Credit Card Fraud Detection/creditcard.csv')
df.head(3)
```

```python
plt.figure(figsize=(8, 4))
plt.title('Repartition before SMOTE')
plt.scatter(df[df["Class"]==1]["Amount"], df[df["Class"]==1]["Class"], label='class 1')
plt.scatter(df[df["Class"]==0]["Amount"], df[df["Class"]==0]["Class"], label='class 0')
plt.legend()
plt.grid(False)
plt.show()
```

```python
plt.figure(figsize = (5, 4))
sns.countplot(df['Class'])
plt.show()
```

```python
y = df["Class"]
X = df.loc[:,df.columns!="label"]
```

```python
smt = SMOTE()
X_smote, y_smote = smt.fit_resample(X, y)
```

```python
plt.figure(figsize = (5, 4))
sns.countplot(y_smote)
plt.show()
```

```python
plt.figure(figsize=(8, 4))
plt.title('Repartition after SMOTE')
plt.scatter(X_smote[y_smote==1]["Amount"], X_smote[y_smote==1]["Class"], label='class 1')
plt.scatter(X_smote[y_smote==0]["Amount"], X_smote[y_smote==0]["Class"], label='class 0')
plt.legend()
plt.grid(False)
plt.show()
```

```python
print("Count two Classes after smote:    ",Counter(y_smote))
```

<h3>Python Implimintation:</h3> (Mammography dataset)

```python
from imblearn.datasets import fetch_datasets
df1 = fetch_datasets()["mammography"]
X, y = df1.data, df1.target
```

```python
y
```

```python
plt.figure(figsize=(8, 4))
plt.title('Repartition Befor BorderLine-SMOTE')
plt.scatter(X[y==-1], X[y==-1], label='class -1')
plt.scatter(X[y==1], X[y==1], label='class 1')
plt.legend()
plt.grid(False)
plt.show()
```

```python
plt.figure(figsize = (5, 4))
sns.countplot(y)
plt.show()
```

```python
from collections import Counter
from imblearn.over_sampling import BorderlineSMOTE
smt = BorderlineSMOTE(random_state=42)
X_res, y_res = smt.fit_resample(X, y)
```

```python
print('Resampled dataset shape %s' % Counter(y_res))
```

```python
plt.figure(figsize=(8, 4))
plt.title('Repartition After BorderLine-SMOTE')
plt.scatter(X_res[y_res==-1], X_res[y_res==-1], label='class -1')
plt.scatter(X_res[y_res==1], X_res[y_res==1], label='class 1')
plt.legend()
plt.grid(False)
plt.show()
```

```python
plt.figure(figsize = (5, 4))
sns.countplot(y_res)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, stratify=y_res, random_state=0)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=30)
knn.fit(X_train, y_train)
y_pred_knn = knn.predict(X_test)

# Predict on dataset which model has not seen before
print(knn.predict(X_test))
```

```python
# Calculate the accuracy of the model
print(knn.score(X_test, y_test))
```

```python
print()
print("ACCURACY OF THE MODEL KNN: ", metrics.accuracy_score(y_test, y_pred_knn))
```
